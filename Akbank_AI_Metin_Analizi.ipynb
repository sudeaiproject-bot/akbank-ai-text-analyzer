```python
from getpass import getpass
import os, google.generativeai as genai

api_key = getpass('Google API Key (gizli): ')
os.environ["GOOGLE_API_KEY"] = api_key
genai.configure(api_key=api_key)
```

```python
model = genai.GenerativeModel("gemini-2.5-flash")
print(model.generate_content("Merhaba! Test için kısa bir cümle oluştur.").text)
```

```python
import gradio as gr

def analiz_et(metin):
    ozet   = model.generate_content(f"Metni 2-3 cümlede tarafsız özetle:\n\n{metin}").text
    baslik = model.generate_content(f"Bu metin için 5 kelimeyi geçmeyen ilgi çekici başlık yaz:\n\n{metin}").text
    anahtar= model.generate_content(f"Bu metinden 5-8 anahtar kelime çıkar:\n\n{metin}").text
    return ozet, baslik, anahtar

arayuz = gr.Interface(
    fn=analiz_et,
    inputs=gr.Textbox(lines=6, label="Metni buraya yapıştır"),
    outputs=[gr.Textbox(label="Özet"),
             gr.Textbox(label="Başlık"),
             gr.Textbox(label="Anahtar Kelimeler")],
    title="Akbank AI – Metin Analizi Aracı",
    description="Yapay zeka destekli metin analizi (özet, başlık, anahtar kelime çıkarımı)"
)
arayuz.launch(debug=True, share=False)
```
