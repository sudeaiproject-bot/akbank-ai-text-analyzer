```python
import os
import google.generativeai as genai
import gradio as gr

os.environ["GOOGLE_API_KEY"] = "AIzaSyBtBDRwIv17IMUgFZST9gp4tWAbQkO8Y7E"
genai.configure(api_key=os.environ["GOOGLE_API_KEY"])

MODEL_NAME = "gemini-2.5-flash"
model = genai.GenerativeModel(MODEL_NAME)

def analiz_et(metin):
    ozet = model.generate_content(f"Metni 2-3 cümlede tarafsız bir dille özetle:\n\n{metin}").text
    baslik = model.generate_content(f"Bu metin için 5 kelimeyi geçmeyen ilgi çekici bir başlık yaz:\n\n{metin}").text
    anahtar = model.generate_content(f"Bu metinden 5-8 adet anahtar kelime çıkar:\n\n{metin}").text
    return ozet, baslik, anahtar

arayuz = gr.Interface(
    fn=analiz_et,
    inputs=gr.Textbox(lines=6, label="Metni buraya yapıştır"),
    outputs=[
        gr.Textbox(label="Özet"),
        gr.Textbox(label="Başlık"),
        gr.Textbox(label="Anahtar Kelimeler")
    ],
    title="Akbank AI – Metin Analizi Aracı",
    description="Yapay zeka destekli metin analizi (özet, başlık ve anahtar kelime çıkarımı)"
)

arayuz.launch(debug=True, share=False)
```
